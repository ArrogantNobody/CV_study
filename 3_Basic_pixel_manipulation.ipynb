```
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```
img = cv2.imread('/content/drive/MyDrive/CV_study/images/banfu.jpg')
px = img[50,50]
print(px)
px = img[50,50,0]#获取blue值
print(px)
px = img[50,50,1]#获取green值
print(px)
px = img[50,50,2]#获取red值
print(px)

img[50,50] = [255,255,255]
print(img[50,50])
```

This kind of pixel operation could be time consuming, use matrix instead. 
array.item() and array.itemset()

```
print(img[10,10])
sca=img.item(10,10,0)#获取bgr值
print(sca)
sca=img.item(10,10,1)
print(sca)
sca=img.item(10,10,2)
print(sca)

img.itemset((10,10,2),100)#给像素赋值
```
