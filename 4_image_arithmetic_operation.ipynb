```
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```
img1 = cv2.imread('/content/drive/MyDrive/CV_study/images/banfu.jpg')
img2 = cv2.imread('/content/drive/MyDrive/CV_study/images/stone.jpg')
```

```
res = cv2.add(img1,img2)
cv2_imshow(res)
```

```
res=cv2.addWeighted(img1,0.7,img2,0.3,0)
cv2_imshow(res)
```
